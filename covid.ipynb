## 2. Number of persons state wise vaccinated for first dose in India 

```python
first_dose_statewise = df.groupby('State')['First Dose Administered'].sum()
print("\n 2. Number of Persons State-wise Vaccinated for First Dose in India:")
print(first_dose_statewise)
```

## 3. Number of persons state wise vaccinated for second dose in India 

```python
second_dose_statewise = df.groupby('State')['Second Dose Administered'].sum()
print("\n3. Number of Persons State-wise Vaccinated for Second Dose in India:")
print(second_dose_statewise)
```

## 4. Number of Males vaccinated

```python
males_vaccinated = df['Male(Individuals Vaccinated)'].sum()
print("\n 4. Number of Males Vaccinated:", males_vaccinated)
```

## 5.Number of Females vaccinated

```python
females_vaccinated = df['Female(Individuals Vaccinated)'].sum()
print("\n 5. Number of Females Vaccinated:", females_vaccinated)
```
